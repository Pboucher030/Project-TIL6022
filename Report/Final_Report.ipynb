# 1. Introduction
Over the past decade, remote work has gained traction in the Netherlands, reshaping mobility and commuting patterns in the country. This research investigates the impact of increased remote work on various mobility trends, transportation choices, and their potential implications for urban planning and environmental sustainability. The study spans a ten-year period from 2013 to 2023, with a particular focus on the acceleration of remote work due to the COVID-19 pandemic. By analyzing data on commuting frequency, transport mode preferences, and emissions, this research aims to quantify the effects of remote work on travel demand, public transportation usage, and environmental impact, ultimately providing insights into how remote work has become an integral part of workplace culture in the Netherlands.


# 2. Main Research Objective

The primary objective of this research is to quantify and analyze how the increase in remote work over the past decade has influenced mobility patterns, transport mode usage, and environmental impact in the Netherlands.


# 3. Research questions
To address the main objective of understanding the impact of increased remote work on mobility in the Netherlands, the research has been divided into four key areas, each focusing on different aspects of this impact. Each area is represented by a research question aimed at examining specific changes over the past decade.

- **RQ1**: How has the prevalence of remote work changed in the Netherlands over the past 10 years?
- **RQ2**: How has the travel motive developed over the last 10 years?
- **RQ3**: To what extent has remote work impacted the use of various transport modes over the past decade?
- **RQ4**: How has the rise in remote work influenced transport-related emissions and air quality in the Netherlands?

These questions together provide a comprehensive framework for analyzing the multifaceted effects of remote work on transportation and environmental sustainability.

### RQ3: To what extent has remote work impacted the use of various transport modes over the past decade? 

This chapter addresses the following research question: To what extent has the shift to remote work affected the use of different transport modes over the past decade? The analysis was conducted in two main stages. The first stage of the analysis involved examining recent trends in the usage of different transport modes in order to gain insight into how they have evolved over time. This provided a foundation for understanding the shifts in transportation preferences.

In the second stage, remote work data was integrated with transport mode usage data for overlapping years, enabling a comparative analysis. Scatter plots were created to illustrate the relationship between the percentage of remote work and the usage of each transport mode, with each data point representing a specific year. The objective was to identify any patterns or correlations between the rise in remote work and changes in transportation choices. 

#### Transportation modes over time
The bar chart provides an overview of the changes in transportation mode usage from 1999 to 2023, illustrating the percentage distribution of different modes for each period.



*Figure [6]: Percentage transportation modes over time*

The bar chart highlights key shifts in transportation modes over time. Car usage as a driver has remained relatively stable, though it shows a slight decrease from 31.9% in 2020 to 30.6% in 2023. Car passenger usage has seen a more notable decline, from 14.1% in 1999 to 10.5% in 2023, indicating reduced shared car trips. Train usage has generally remained low but experienced a sharp drop after 2020, reaching 2.4% in 2021 and only partially recovering to 3.4% by 2023. Bus/tram/metro usage also decreased gradually, from 3.5% in the early 2000s to 3.1% in 2023. Walking is one of the few modes that has increased, rising from 19.1% in 1999 to 23.5% in 2023, particularly since 2020. Cycling has remained popular and stable, consistently around 26-28%, while other modes have retained a small, steady share of around 3-4%.

Trends and implications
- Decreased use of shared and public transportation: The decline in car passenger, train, and bus/tram/metro usage suggests a reduced reliance on shared and public transport, especially in recent years. This trend may be linked to remote work, which reduces the need for commuting and impacts public transport use.
- Increased walking: Walking has seen a steady increase, potentially as people engage in more local activities or short-distance travel near their homes.
- Stability in cycling and car driving: Despite some fluctuations, cycling and car driver usage have remained relatively stable, indicating these modes are resilient to recent changes in lifestyle and work habits.

#### Relationship between remote work and transport modes
The scatter plots provide a visual representation of the relationship between remote work percentages and the usage percentages of various transport modes from 2015 to 2023. Each point on the plots represents a specific year, allowing us to observe trends and correlations between increased remote work and transport choices.



*Figure [7]: Relation between remote work and car (driver)*

Car usage as a driver shows a clear negative correlation, with usage decreasing from around 31.75% in 2015 to approximately 30.61% in 2023. This suggests that higher levels of remote work reduce the need for individuals to commute by car, leading to a gradual decline in car usage by drivers.



*Figure [8]: Relation between remote work and car (passenger)*

Similarly, car usage as a passenger also demonstrates a negative trend, with a decline from 11.22% in 2015 to 10.54% in 2023. This suggests that as remote work reduces the need for shared commutes, fewer people are traveling as car passengers.



*Figure [9]: Relation between remote work and train*

Train usage also shows a negative correlation with remote work. As remote work percentages rise, train usage drops from around 2.64% in 2015 to 2.38% in 2023, indicating a reduced dependency on public transit options like trains due to decreased commuting needs.



*Figure [10]: Relation between remote work and bus/tram/metro*

The pattern for bus, tram, and metro usage aligns with this trend. Usage of these modes declines from 3.96% in 2015 to 3.40% in 2023 as remote work becomes more prevalent, reflecting a reduced reliance on urban public transit with increased remote work.



*Figure [11]: Relation between remote work and walking*

Walking is the only mode that shows a positive correlation with remote work. As remote work percentages increase, walking usage rises from 19.80% in 2015 to 23.47% in 2023. This may be due to more localized movement patterns, as people working from home tend to take more short trips on foot within their neighborhoods.



*Figure [12]: Relation between remote work and biking*

Cycling shows a less pronounced relationship with remote work, though there is a slight negative trend. Cycling usage decreases marginally from 27.72% in 2015 to 26.53% in 2023, indicating that while cycling remains a popular mode of transport, its usage may be influenced less by remote work and more by other factors.


Overall, the scatter plots reveal a consistent trend: as remote work increases, the usage of most transportation modes, especially public transit and shared car trips, decreases. The only exception is walking, which shows a positive correlation, likely due to more localized travel patterns associated with working from home. This analysis highlights the impact of remote work on reducing reliance on conventional commuting modes and promoting more pedestrian-friendly travel habits.

The analysis reveals that the increase in remote work over the past decade has influenced the use of various transport modes. As remote work has risen, there has been a notable decline in public transit (train, bus/tram/metro) and shared car trips, as well as a slight reduction in car usage as a driver. Walking, on the other hand, has increased, likely due to more localized travel patterns associated with working from home, as people generally do not commute to work on foot over long distances. Cycling has shown only a slight decrease, indicating that it may be less affected by remote work trends. These findings suggest that remote work reduces the reliance on traditional commuting modes and encourages more pedestrian-friendly, local travel.


While the observed correlations between remote work and transportation modes are evident, it is essential to consider additional factors. The impact of the COVID-19 pandemic likely influenced these trends, as it led to a temporary reduction in the use of public transit and shared transportation options. However, there are signs of recovery, with many transportation modes beginning to rebound as restrictions ease and commuting resumes, suggesting that some declines may be due to pandemic-related factors rather than remote work alone.

Moreover, the analysis was limited to only six data points, as only a few years had overlapping data for both remote work percentages and transport mode usage. This limited dataset restricts the strength of the correlations observed. In the future, as more data becomes available, a more comprehensive analysis with additional years and data points could provide a clearer understanding of the long-term relationship between remote work and transportation preferences. This would be particularly valuable for cities and policymakers in developing resilient, sustainable mobility strategies that accommodate evolving work and travel patterns.



### RQ4: How has the rise in remote work influenced transport-related emissions and air quality in the Netherlands?

Figure 13 illustrates the trend in total CO2 emissions in the Netherlands over time, peaking in the early 2000s, gradually declining, and then sharply dropping around 2020. This reduction coincides with the COVID-19 pandemic, during which commuting and overall transport activity were significantly curtailed. Figure 14 breaks down emissions by transport mode, highlighting that private cars consistently contribute the most emissions, followed by motorcycles and commercial vehicles. In contrast, emissions from commercial vehicles remained steady, as companies continued transporting goods despite pandemic restrictions.

In contrast, emissions from commercial vehicles do not show a similar drop. This can be attributed to the continued need for companies to transport goods, which remained unaffected by the shift to remote work

# Conclusion

The research has shown that the increase in remote work over the past decade has influenced not only commuting patterns and transport mode usage but also contributed to a reduction in transport-related emissions and an improvement in air quality in the Netherlands. The shift to remote work, accelerated by the COVID-19 pandemic, led to a significant drop in CO2 emissions around 2020 due to reduced commuting and road traffic. This trend in emissions has not fully rebounded to pre-pandemic levels, suggesting that sustained remote work has a lasting effect on lowering overall emissions. Private cars, the largest contributors to transport emissions, showed a notable decline in emissions from 2020 onward, while commercial vehicle emissions remained steady, as essential goods transport continued unaffected. These findings highlight that remote work adoption may play a crucial role in achieving long-term sustainability and emissions reduction goals.

# Discussion
While the research offers valuable insights into the effects of remote work on mobility and environmental factors, several considerations limit the findings. First, the COVID-19 pandemic acted as a significant external influence, particularly on public transport usage, as restrictions led to temporary declines that may not solely be attributable to remote work. As commuting patterns begin to stabilize post-pandemic, the relative influence of remote work versus pandemic-induced changes remains partially uncertain.

Additionally, the limited data points for transport mode usage in relation to remote work restrict the analysis, as only a few years provide overlapping data on both metrics. This limitation affects the robustness of the correlations drawn in the study. Future research, with a broader dataset covering more years, could enhance understanding of the long-term effects of remote work on travel behavior and support policymakers in developing resilient and sustainable mobility strategies that accommodate shifting work and travel patterns.

To enhance future studies, collecting consistent, annual data on remote work and transportation choices would improve the accuracy of trend analysis. Focusing on regional differences could reveal how urban and rural areas adapt differently to increased remote work. Additionally, incorporating qualitative insights, such as surveys or interviews, could provide context on why people choose specific transport modes when working remotely. Exploring the impact of hybrid work models would also be valuable, as these arrangements continue to shape commuting patterns. These steps would support more targeted, resilient mobility strategies that align with changing work habits.
