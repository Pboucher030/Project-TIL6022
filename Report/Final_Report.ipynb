# 1. Introduction
Over the past decade, remote work has gained traction in the Netherlands, reshaping mobility and commuting patterns in the country. This research investigates the impact of increased remote work on various mobility trends, transportation choices, and their potential implications for urban planning and environmental sustainability. The study spans a ten-year period from 2013 to 2023, with a particular focus on the acceleration of remote work due to the COVID-19 pandemic. By analyzing data on commuting frequency, transport mode preferences, and emissions, this research aims to quantify the effects of remote work on travel demand, public transportation usage, and environmental impact, ultimately providing insights into how remote work has become an integral part of workplace culture in the Netherlands.


# 2. Main Research Objective

The primary objective of this research is to quantify and analyze how the increase in remote work over the past decade has influenced mobility patterns, transport mode usage, and environmental impact in the Netherlands.


# 3. Research questions
To address the main objective of understanding the impact of increased remote work on mobility in the Netherlands, the research has been divided into four key areas, each focusing on different aspects of this impact. Each area is represented by a research question aimed at examining specific changes over the past decade.

- **RQ1**: How has the prevalence of remote work changed in the Netherlands over the past 10 years?
- **RQ2**: How has the travel motive developed over the last 10 years?
- **RQ3**: To what extent has remote work impacted the use of various transport modes over the past decade?
- **RQ4**: How has the rise in remote work influenced transport-related emissions and air quality in the Netherlands?

These questions together provide a comprehensive framework for analyzing the multifaceted effects of remote work on transportation and environmental sustainability.

### RQ1: How has the prevalence of remote work changed in the Netherlands over the past 10 years?
The first research question explains how the prevalence of remote work in the Netherlands has changed over the last 10 years. This gives a good insight into the trend of working from home and how this will develop in the future years. Figure [1] shows the trend line of the number of people working from home in 2013 until 2023. The group of people working from home is defined as people who are always working from home and people who mostly or sometimes work from home. As can be seen in the figure, when the time passes more people are starting to work from home, with a clear increase in 2020. This increase is a result of the Covid crisis which started during this period (Rijksoverheid, 2024). Because people were not able to work in an office anymore due to the 1.5 meter restriction, they had to find another way to continue their work. That is why people started to work from home as can be seen in the figure. Following the spike in 2020, a significant proportion of employees have continued to work from home even after the conclusion of the crisis. The transition to remote work has had a significant and enduring impact, reducing the necessity for daily commuting for a considerable number of employees. It is anticipated that this reduction in physical commutes will have a considerable impact on overall travel demand and mobility patterns in the Netherlands, influencing both the use of public transportation and private car travel.



*Figure [1]: Number of people working from home 2013 until 2023*

In figure [2] the type of workers over the last 10 years is shown. The graph shows trends in remote working among workers, segmented by those who often work from home, sometimes work from home, and do not work from home. Over time, there has been a noticeable increase in both the number of people who sometimes and often work from home, particularly after 2020. This suggests a shift towards more flexible work arrangements, influenced by the COVID-19 pandemic, which led to a decline in the proportion of people who do not work from home.



*Figure [2]: Types of remote workers 2013 until 2023*

Figure [3] illustrates the percentage distribution of remote work from 2013 to 2023. It shows a steady increase in the proportion of people who sometimes work from home and a notable rise in the year 2020 where it rises from 23.1% to 33.3% This increase develops in the years after. Consequently, the percentage of workers who do not work from home has declined significantly, from about 65% in 2014 to around 48% in 2022. This shift highlights the growing trend towards flexible work arrangements, likely accelerated by the pandemic starting in 2020.



*Figure [3]: Types of remote workers 2013 until 2023 expressed in percentages*

Looking at the figures, it can be concluded that remote working had already been increasing over the past years, with COVID-19 accelerating this trend. The data shows that the category "Sometimes working from home" has grown significantly, indicating that working from home occasionally has become more normalized and is increasingly becoming part of workplace culture.
